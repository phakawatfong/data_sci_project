```python
data = driver.page_source
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(data)
```

```python
driver.execute_script("document.body.style.zoom='100%'")
```

```python
len(soup.find_all('div', {'class':'VTjd7p whIxGK'}))
```

```python
all_product = soup.find_all('div', {'class':'VTjd7p whIxGK'})
```

```python
for product in all_product:
    print(product.text)
    print(' ')
```

```python
product_list = []
for product in all_product:
    product_list.append(product.text)
    
product_list[1:3]
```

```python
all_price = soup.find_all('div', {'class':'vioxXd rVLWG6'})
```

```python
all_price[0].text
```

```python
price_list = []
for price in all_price:
    price_list.append(price.text)
    
price_list[1:3]
```

```python
all_sales = soup.find_all('div', {'class': 'r6HknA uEPGHT'})
all_sales[0].text
```

```python
sales_list = []
for sale in all_sales:
    sales_list.append(sale.text)

sales_list[1:3]   
```

```python
print(len(product_list))
print(len(price_list))
print(len(sales_list))
```

```python
import pandas as pd
```

```python
shopee_df = pd.DataFrame([product_list,price_list,sales_list])
```

```python
shopee_df = shopee_df.transpose()
# shopee_df.columns = ['product name','price','total sales']
```

```python
shopee_df.columns = ['product name','price','total sales']
```
